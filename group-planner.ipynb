```python
import random 
```

```python
# Person Class
class Person:
    def __init__(self, ind, gender, acq_ind_list): 
        self.ind = ind 
        self.gender = gender
        self.acq_ind_list = acq_ind_list
```

```python
# Create a list of Person
num_person = 24
list_person = []
LABEL_MALE = 1
LABEL_FEMALE = 0

for i in range(num_person):
    list_person.append(Person(i,i%2,[]))
```

```python
NUM_GROUP = 6
MAX_PEOPLE_IN_GROUP = 4
MAX_STAGE = 4
# 6 x 6 x 4
list_turn_stage = [[[],[],[],[],[],[]]]*MAX_STAGE
print(list_turn_stage)
```

```python
def no_acq(person, list_person_ind_in_group):
    
    #print("SET",person.acq_ind_list,list_person_ind_in_group)
    
    return (set(person.acq_ind_list)&set(list_person_ind_in_group) == set())
```

```python
def no_gender_imbalance(person, list_person_ind_in_group, max_people_in_group):
    
    list_person_gender_in_group_new = [list_person[person_ind].gender for person_ind in list_person_ind_in_group] + [person.gender]
    
    return (list_person_gender_in_group_new.count(LABEL_MALE) <= max_people_in_group/2) and (list_person_gender_in_group_new.count(LABEL_FEMALE) <= max_people_in_group/2)
```

```python
# Arrange the first round. Allocate Person into 6 group of 4.
def allocate_group(stage_num, bool_first_trial):
    
    if bool_first_trial:
        print("STAGE NUM", stage_num)
    
    # Make it empty at first
    for group_ind in range(NUM_GROUP):
        list_turn_stage[stage_num][group_ind] = []
    
    # random.sample() makes it RANDOM
    for person in random.sample(list_person,len(list_person)):
            
        for group_ind in range(NUM_GROUP):
            
            if len(list_turn_stage[stage_num][group_ind]) >= MAX_PEOPLE_IN_GROUP:
                continue
            
            # If Person not in group
            if person.ind not in list_turn_stage[stage_num][group_ind]:
                
                #print(list_turn_stage[stage_num])
                
                # If Person has no acq in this group, add it.
                if no_acq(person, list_turn_stage[stage_num][group_ind]) and no_gender_imbalance(person, list_turn_stage[stage_num][group_ind], MAX_PEOPLE_IN_GROUP):
                    
                    list_turn_stage[stage_num][group_ind].append(person.ind)
                    
                    # print("ADD PEOPLE", person.ind, "TO GROUP", group_ind)
                    
                    break
                else:
                    continue
```

```python
def update_acq_list(stage_num):
    
    for group in list_turn_stage[stage_num]:
        
        print("GROUP:", group)
        
        for x, people_ind in enumerate(group):
            
            people = list_person[people_ind]
            people.acq_ind_list = people.acq_ind_list + group
```

```python
def flatten(t):
    return [item for sublist in t for item in sublist]
```

```python
trial_num = 0

for stage_ind in range(MAX_STAGE):
    allocate_group(stage_ind,True)
    
    # Check if successfully arrange ALL
    while(len(flatten(list_turn_stage[stage_ind])) != MAX_PEOPLE_IN_GROUP*NUM_GROUP):
        
        
        trial_num = trial_num + 1
        # print("TRIAL", trial_num, "--> Matched: ", len(flatten(list_turn_stage[stage_ind])))
        
        allocate_group(stage_ind,False)
        
    
    update_acq_list(stage_ind)
```
